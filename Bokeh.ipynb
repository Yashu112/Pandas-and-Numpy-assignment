Q1. How can you create a Bokeh plot using Python code?

Ans. For creating bokeh plot, we first have to import bokeh.io, bokeh.plotting, figure, output_file and show. We call boleh.io.output_notebook to display plot in Jupyter notebook. Then, we create an html file which shows the output plot. 

The libraries have been imported in the above cell. Below is an example of creating a simple plot using bokeh:

```python
import numpy as np
x = np.linspace(0,10,201)
y = np.tan(x)

output_file=('tan(x).html')
p = figure(title='Plot of tan(x)')
p.xaxis.axis_label='x'
p.yaxis.axis_label='tan(x)'
p.line(x,y)
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans. Glyph is a class of objects in Bokeh library. Glyphs are various types of visual shapes and objects which can be used to reperesent the data points in our plot. Various common glyphs can simply be added as a method of the figure object. 

Example is as shown:

```python
#Solution-2
from bokeh.layouts import row

x=[1,2,3,4,]
y=[1,2,3,4,]

pl1 = figure(title='GLYPHS',width=250, height=250)
pl1.star(x,y,size=25)

pl2 = figure(title='GLYPHS', width=250, height=250)
pl2.annulus(x,y,inner_radius=.1, outer_radius=.2)

pl3 = figure(title='GLYPHS', width=250, height=250)
pl3.ellipse(x,y, width=.5, height=.2, color='red')

pl4 = figure(title='GLYPHS', width=250, height=250)
pl4.triangle(x,y, size=15, color='green')

show(row(pl1,pl2,pl3,pl4))
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Ans. First we create a plot object. Then using methods such as axes, title, etc, and attributes, we can customize the appearance of plot. 

An example using flowers dataset is shown as follows:

```python
from bokeh.sampledata.iris import flowers
flowers.head()
```

```python
p=figure(title='flowers')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.background_fill_color = 'pink'
p.circle(flowers['petal_length'], flowers['petal_width'], fill_color='red', legend_label = 'red points', size=10)
show(p)
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

Ans. We can embed a Bokeh plot into a web page or dashboard using Flask in the following way

```python
from flask import Flask, render_template
from bokeh.plotting import figure, show
from bokeh.resources import CDN
from bokeh.embed import file_html

app = Flask(__name__)

def create_bokeh_plot():
    plot = figure(title="Bokeh Plot Example", width=600, height=600)
    plot.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5])  
    html = file_html(plot, CDN, "my plot")
    return html

@app.route('/')
def index():
    plot = create_bokeh_plot()
    return plot

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
