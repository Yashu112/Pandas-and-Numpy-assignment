Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans. 5 plots that can be plotted using Matplotlib are as follows:
1. line plot - it plots a given set of x and y co-ordinates in the form of a line
2. scatter plot - it plots a set of points which are scattered in the x, y co-ordinates
3. bar plot - A bar plot shows data in the form of bars, vertical or horizontal. In a bar plot, one axis shows numerical data and the other axis generally shows the different categories.
4. histogram/distplot - Distplot stands for distribution plot which is anothter name given to histogram plot. In a distribution plot, as the name suggests, the frequency distribution of data, the value counts, is plotted. It shows how many times a particular value of data occurs.
5. pie chart - it shows the fraction of different categories of data in the total data.

One example of each is as shown:

```python
#1 line plot:
x = np.linspace(0,10,201)
y = np.sin(x)
plt.title('Line Plot')
plt.plot(x,y,)
plt.xlabel("X-axis")
plt.ylabel("Y-axis (sin(x))")
plt.show()
```

```python
#2 scatter plot:
import numpy as np
x = np.random.randint(1,10,50)
y = np.random.randint(1,10,50)
plt.title('Scatter Plot')
plt.scatter(x,y)
plt.show()
```

```python
#3 Bar plot
x = ['A','B','C','D','E']
y = [20,100,40,60,70]
plt.title('Bar Plot')
plt.bar(x,y)
plt.show()
```

```python
#4 Histogram 
x = np.random.randint(1,10,50)
plt.title('Histogram/Distplot')
plt.hist(x)
plt.show()
```

```python
#pie chart
x = [20,100,40,60,70]
plt.title('Pie Chart')
plt.pie(x, labels=['A','B','C','D','E'])
plt.show()
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Note: Also add title, xlabel, and ylabel to the plot.


```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
#Solution-2
plt.scatter(x,y)
plt.xlabel("This is X-axis")
plt.ylabel("This is Y-axis")
plt.title("TTITLE - SCATTER PLOT")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

Ans. The subplot() function is used to plot multiple plots in one figure. The four lines given in question are plotted as shown:

```python
import numpy as np
#For line 1: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,2,1)
plt.plot(x,y)

#For line 2: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,2,2)
plt.plot(x,y)

#For line 3: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,2,3)
plt.plot(x,y)

#For line 4: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)
plt.plot(x,y)

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Ans. A bar plot shows data in the form of bars, vertical or horizontal. In a bar plot, one axis shows numerical data and the other axis generally shows the different categories.

The bar plot is generally used to plot numerical data of different categories.

The data in question is plotted as shown:

```python
plt.bar(company,profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

Ans. A box plot shows distribution of data in the form of box.
It's components from bottom to top are - 
- the minimum value, excluding the outliers
- the first quartile, Q1. 25% of data lies below Q1
- the median line, Q2
- the third quartile, Q3. 25% of data lies above Q3
- the maximum value, excluding the outliers

The plots of data given are as shown:

```python
#Solution-5
plt.boxplot([box1,box2])
plt.show()
```
