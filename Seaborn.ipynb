Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans. 5 plots that can be plotted using Seaborn are as follows:
1. line plot - it plots a given set of x and y co-ordinates in the form of a line. It is used to draw a line which shows the variation of a particular dependent variable with respect to an independent variable. It is helpful for finding the nature of relation between the two variables, if it is linear, exponential, etc.

2. scatter plot - it plots a set of points which are scattered in the x, y co-ordinates. It is used to understand, variation of random data. It gives u

3. bar plot - A bar plot shows data in the form of bars, vertical or horizontal. In a bar plot, one axis shows numerical data and the other axis generally shows the different categories.

4. histogram/distplot - Distplot stands for distribution plot which is anothter name given to histogram plot. In a distribution plot, as the name suggests, the frequency distribution of data, the value counts, is plotted. It shows how many times a particular value of data occurs.

5. box plot - it shows distribution of data in the form of box. It's components from bottom to top are - the minimum value, the first quartile, Q1 (25% of data lies below Q1), the median line, Q2, the third quartile, Q3(25% of data lies above Q3), the maximum value, excluding the outliers. It is used to provide a quick visual summary of data.

Simple examples of each type of plot:

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
#line plot
import numpy as np
x = np.linspace(0,10,201)
y = np.exp(x)
lp = sns.lineplot(x=x, y=y)
lp.figure.suptitle('Line Plot')
lp.figure.supxlabel('x')
lp.figure.supylabel('log(x)')
```

```python
#scatter plot
scat_plot = sns.scatterplot(x=tip.total_bill, y=tip.tip, hue=tip.day)
scat_plot.figure.suptitle('Scatter Plot')
```

```python
#bar plot
x = ['A','B','C','D','E']
y = [20,100,40,60,70]
sns.barplot(x=x,y=y).set(title='Bar Plot ')
```

```python
#histogram
sns.histplot(x=tip.tip, hue=tip.day).set(title='Histogram')
```

```python
#boxplot
sns.boxplot(x=tip.day, y=tip.total_bill).set(title='Box Plot')
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
#Solution-2
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue='event')
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue='event')
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue='event', style='region')
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

Ans. The required plot is as shown -

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
